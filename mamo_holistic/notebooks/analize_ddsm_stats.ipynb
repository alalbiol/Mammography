```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

En este script analizo el resultado del scrip check_ddsm_images.py

Basicamente miro los valores maximo y minimo media y demas para ver si hay imagenens malas

```python
ddsm_stats = pd.read_csv('../scripts/ddsm_images_stats.csv')
```

```python
ddsm_stats.head()
```

```python
ddsm_stats_sorted = ddsm_stats.sort_values(by='image_contrast', ascending=True)
ddsm_stats_sorted.head()
```

```python
len(ddsm_stats_sorted.loc[ddsm_stats_sorted['image_contrast'] < 30000])

for k, row in ddsm_stats_sorted.loc[ddsm_stats_sorted['image_contrast'] < 1000].iterrows():
    print("Image: ", row['image_path'], "max: ", row['image_max'], "min: ", row['image_min'], "contrast: ", row['image_contrast'])
```

```python
ddsm_stats_sorted = ddsm_stats.sort_values(by='image_std', ascending=True)
ddsm_stats_sorted.head(10)
```

```python
for k, row in ddsm_stats_sorted.loc[ddsm_stats_sorted['image_std'] < 10000].iterrows():
    print("Image: ", row['image_path'], "max: ", row['image_max'], "min: ", row['image_min'], "contrast: ", row['image_std'])
```
