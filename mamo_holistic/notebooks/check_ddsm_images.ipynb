# Check DDSM_png_16bit_1152x896

```python
root_folder = pathlib.Path('/home/alalbiol/Data/mamo/DDSM_png_16bit_1152x896')

all_mask_files = [ f for f  in root_folder.glob('**/*.png') if 'mask_' in f.name ]
print("Number of mask files: ", len(all_mask_files))
```

```python
def maskname2imagenames(mask_filename):
    maskname = mask_filename.name
    maskname_parts = maskname.split('_')
    image_name = "_".join(maskname_parts[:4]) + '.png'
    return mask_filename.parent / image_name

sample_mask = all_mask_files[0]
sample_image = maskname2imagenames(sample_mask)

print("Sample mask: ", sample_mask)
print("Sample image: ", sample_image)

    
```

```python
fig, ax = plt.subplots(3,2, figsize=(10,15))

ax = ax.ravel()

for k in range(6):
    rand_ind = np.random.randint(len(all_mask_files))
    sample_mask = all_mask_files[rand_ind]
    sample_image = maskname2imagenames(sample_mask)
    show_mask_image(sample_image, sample_mask, ax[k])
    
```
