# VISUALIZAR IMÉNES EN CUADRÍCULA 

```python
# copilot

fig, axs = plt.subplots(2, 2, figsize=(10, 15))

for i, ax in enumerate(axs.flat):
    img = mpimg.imread(images[i])
    ax.imshow(img, cmap='gray')
    ax.set_title(f'IMamografia {i+1}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```
